```python
class Node(object):
    def __init__(self,key):
        self.key = key
        self.parent = self.left = self.right = None
```

```python
class BST(object):
    def __init__(self):
        self.root = None
        self.size = 0

    def __len__(self):
        return self.size

    def traversal(self,mod):
        if mod == 'pre':
            self._preorder(self.root)
        elif mod == 'ino':
            self._inorder(self.root)
        else:
            self._postorder(self.root)

    # BST에 해당 key값을 가지는 Node를 삽입
    # 동일한 key는 무시가 되고, root일 경우 or parent의 key와 key가 다를 경우만 insert를 진행
    def insert(self,key):
        new_node = Node(key)
        parent = self._find_loc(key,self.root,None)

        if (parent == None) or (parent.key != key):
            if parent == None:          # parent == None -> 이 Node가 root node
                self.root = new_node
            else:                       # parent != None, parent.key != key -> parent와 동일한 key 값이 아님으로 왼쪽인지 오른쪽인지 구분하여 삽입
                new_node.parent = parent
                if parent.key < key:
                    parent.right = new_node
                else:
                    parent.left = new_node
            
            self.size += 1
            return new_node
        else:
            print("key is already insert")
            return parent

    def delete(self,key,mod):
        curr = self._find_loc(key,self.root,None)
        if curr.key == key:
            if mod =='Merging':
                self._deleteByMerging(curr)
            elif mod == 'Copying':
                self._deleteByCopying(curr)
   
    def IsContains(self,key):
        res = self._find_loc(key,self.root,None)

        if res is not None:
            return (res.key == key)
        else:
            print("This BST is Empty")
            return None

    def _preorder(self,parent):
        if parent is not None:
            print(parent.key,end=' ')
            
            if parent.left is not None:
                self._preorder(parent.left)
                
            if parent.right is not None:
                self._preorder(parent.right)
        
    def _inorder(self,parent):
        if parent is not None:
            if parent.left is not None:
                self._inorder(parent.left)

            print(parent.key,end=' ')

            if parent.right is not None:
                self._inorder(parent.right)
        
    def _postorder(self,parent):
        if parent.left is not None:
                self._postorder(parent.left)

        if parent.right is not None:
                self._postorder(parent.right)
    
        print(parent.key,end=' ')
    
    # BST에 들어갈 key값의 위치를 검색
    # key값 node가 있다 = 해당 node return , key값 node가 없다 = key값이 삽입될 parent node return
    def _find_loc(self,key,curr,parent):
        if self.size == 0:
            return None

        if curr is None:
            return parent
        
        if curr.key == key:
            return curr
        elif curr.key < key:
            parent = curr
            curr = curr.right
        else:
            parent = curr
            curr = curr.left
        
        return self._find_loc(key,curr,parent)
        
    def _deleteByMerging(self,curr):
        left = curr.left
        right = curr.right
        parent = curr.parent
        
        if left is not None:
            change = left
            max = left
            while max.right is not None:
                max = max.right

            if right is not None:
                right.parent = max
            max.right = right

        else:
            change = right

        if parent is not None:
#            if change is not None : 
#                change.parent = parent
#
#                if parent.key < change.key:
#                    parent.right = change
#                else:
#                    parent.left = change       
#            else:
#                if parent.key < curr.key:
#                    parent.right = change
#                else:
#                    parent.left = change

            if change is not None : 
                change.parent = parent  

            if parent.right is curr:
                parent.right = change
            else:
                parent.left = change

        else:
            self.root = change
            if change is not None:
                change.parent=None

        self.size -= 1

    # delete node의 left의 max를 copy하는 deleteByCopying code
    def _deleteByCopying(self,curr):
        left, right = curr.left, curr.right
        parent = curr.parent 

        # delete node의 left가 있는 경우 
        if left is not None:
            max = left
            while max.right is not None:
                max = max.right
            
            curr.key = max.key

            if max.left is not None:
                max.left.parent = max.parent
             
            if max is left:
                max.parent.left = max.left
            else:
                max.parent.right = max.left

        # delete node의 right가 있거나 delete node가 Leaf node인 경우 
        else:
            # delete node의 parent가 있는 경우
            if parent is not None:          
                # delete node가 leaf node가 아닌 경우
                if right is not None:       
                    right.parent = parent
                # 
                if parent.left == curr:
                    parent.left = right
                else:
                    parent.right = right
            # delete node가 root인 경우
            else:                           
                self.root = curr.right
                if right is not None:
                    right.parent = None    
        self.size -= 1
```
