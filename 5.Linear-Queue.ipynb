## Linear Queue
Linked List로만 구현한다. 

arraylist로 Linear를 만드는 것은 원소들을 계속 앞으로 당겨줘야하기 때문에 비효율 적이다.

```python
# pointer
class Node(object):
    def __init__(self,data):
        self.data = data
        self.next = None

class Linear_Queue(object):
    def __init__(self):
        self.head=None
        self.tail=None
        self.size = 0

    def enqueue(self,data):
        new_node = Node(data)
        if self.IsEmpty():
            self.tail = new_node
        else:
            self.head.next = new_node
        self.head = new_node
        self.size += 1

    def dequeue(self):
        if self.IsEmpty():
            return "Queue is Empty"
        ret = self.tail
        self.tail = ret.next
        self.size -= 1
        return ret.data
    
    def sizeOf(self):
        return self.size
    
    def IsEmpty(self):
        return self.tail is None
            
    def top(self):
        if self.IsEmpty():
            return None
        return self.tail.data

    def clear(self):
        self.head = None
        self.tail = None
        self.size = 0
```

```python
# dummy
class Node(object):
    def __init__(self,data):
        self.data = data
        self.next = None

class Linear_Queue(object):
    def __init__(self):
        self.head=Node(None)
        self.tail=Node(None)
        self.size = 0

        self.head.next = self.tail
        
    def enqueue(self,data):
        new_node = Node(data)
        self.head.next.next = new_node 
        self.head.next = new_node
        self.size += 1

    def dequeue(self):
        if self.IsEmpty():
            return "Queue is Empty"
        ret = self.tail.next
        self.tail.next = ret.next
        self.size -= 1 
        if self.size is 0:
            self.clear()
        return ret.data
    
    def sizeOf(self):
        return self.size
    
    def IsEmpty(self):
        return self.tail.next is None
        
    def top(self):
        if self.IsEmpty():
            return None
        return self.tail.next.data

    def clear(self):
        self.head.next = self.tail
        self.tail.next=None
        self.size = 0
```

```python
test = Linear_Queue()
for i in range(10):
    test.enqueue(i**2)

print(test.top())
print(test.sizeOf())

for i in range(10):
    print(test.dequeue(), end = ' ')
print()

print(test.sizeOf())

print("===================================")

for i in range(10):
    test.enqueue(i**3)
    
print(test.top())
print(test.sizeOf())

for i in range(10):
    print(test.dequeue(), end = ' ')
print()

print(test.sizeOf())
```
