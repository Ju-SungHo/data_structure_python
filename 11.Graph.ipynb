```python
# adjacency list
class Graph(object):
    def __init__(self,size=6):
        self.G=[[] for i in range(size)]
        self.V = size

    def __len__(self):
        return len(self.G)
    
    def setGraph(self, Edge_list, Vertex):
        if self.V is not Vertex:    
            self.G=[[] for i in range(Vertex)]
        
        for edge in Edge_list:
            self.G[edge[0]].append(edge[1])
        
        self.V = Vertex
        
    def addEdge(self,src,dst):
        self.G[src].append(dst)

    def isConnect(self,src,dst):
        for node in self.G[src]:
            if node is dst:
                return True
        return False

    def printAdjacentVertex(self,Vertex):
        print(f'{Vertex}s Adjacent Vertex : {self.G[Vertex]}')
    
    def deleteEdge(self,src,dst):
        self.G[src].remove(dst)
```

```python
test = Graph(5)

edge = [(0, 1), (1, 0), (1, 3), (3, 1), (3, 6), (6, 3), (0, 2), (2, 0), (2, 4), (4, 2), (2, 5), (5, 2)]
test.setGraph(edge,7)

print(test.G)
test.printAdjacentVertex(0)
```

## DFS : Depth First Search

```python
# recursive (ver. DFS tree & time check)
vname = {0:'a', 1:'b', 2:'c', 3:'d', 4:'e', 5:'f', 6:'g', 7:'h'}
curr_time = 1
Vertex_num = 8

def DFS(v):
    global curr_time            # 방문 순서 check
    mark[v] = "visited"
    pre[v] = curr_time          # v에 방문한 순간 = 탐색 순서
    curr_time += 1

    # V에 인접한 모든 node 방문
    for node in edge[v]:
        if mark[node] is not "visited":
            parent[node] = v    # DFS tree
            DFS(node)

    # V에 인접한 모든 node에 대해 방문을 끝냄
    post[v] = curr_time         # v에서 DFS가 완료된 순간
    curr_time += 1

test=Graph(Vertex_num)

edge = [(0,1),(0,2),(1,0),(1,2),(1,4),(1,7),(2,0),(2,1),(2,3),(3,2),(3,5),(4,1),(4,6),(5,3),(6,4),(7,1)]
test.setGraph(edge,Vertex_num)
edge = test.G 

mark = [None] * Vertex_num
pre = [None] * Vertex_num
parent = [0] * Vertex_num
post = [None] * Vertex_num

DFS(0)

for i in range(len(vname)):
    print(vname[i],end = ' ')
print()
for i in parent:
    print(vname[i],end=' ')
```
