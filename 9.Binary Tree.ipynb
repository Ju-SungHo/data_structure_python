## Binary Tree
표현법 1 : BT tree -> [array , List] => heap  
표현법 2 : Node & Link

```python
class Node(object):
    def __init__(self,key):
        self.key = key
        self.parent=self.left=self.right=None
    
    def __str__(self):
        return str(self.key)
```

```python
class BT(object):
    def __init__(self,root):
        self.root = root
        self.order = []

    def _clear(self):
        if len(self.order) > 0:
            self.order = []

    def traversal(self,mod='pre'):
        self._clear()

        if mod == 'pre':
            self._preorder(self.root)
        elif mod == 'ino':
            self._inorder(self.root)
        else:
            self._postorder(self.root)

    # Node 탐색 -> left sub tree 탐색 -> right sub tree 탐색 
    def _preorder(self,parent):
        if parent is not None:
            self.order.append(parent.key)

            if parent.left is not None:
                self._preorder(parent.left)
                
            if parent.right is not None:
                self._preorder(parent.right)


    # left sub tree 탐색 -> Node 탐색 -> right sub tree 탐색
    def _inorder(self,parent):
        if parent is not None:
            if parent.left is not None:
                self._inorder(parent.left)

            self.order.append(parent.key)

            if parent.right is not None:
                self._inorder(parent.right)

    # left sub tree 탐색 -> right sub tree 탐색 -> Node 탐색
    def _postorder(self,parent):
        if parent.left is not None:
                self._postorder(parent.left)

        if parent.right is not None:
                self._postorder(parent.right)
    
        self.order.append(parent.key)

    
```

```python
a = Node(6)
b = Node(9)
c = Node(1)
d = Node(5)

a.left = b
a.right = c
b.parent = c.parent = a
b.right = d
d.parent = b

BTt = BT(a)

BTt.traversal('pre')
print(f"preorder path : {BTt.order}")

BTt.traversal('ino')
print(f"inorder path : {BTt.order}")

BTt.traversal('post')
print(f"post order path : {BTt.order}")
```
