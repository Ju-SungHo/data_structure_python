```python
class ArrayList():
    def __init__(self,size=10):
        self.pre_index = 0          # arraylist의 현재 end data index
        self.max_size = size        # arraylist의 max size
        self.arrList=[None] * self.max_size 

    def get(self,index):
        if index >= self.max_size or index < 0:
            raise Exception('Invalid index')

        return self.arrList[index] 
    
    def _increase_size(self):
        new_max_size = self.max_size * 2        # 기본 max size의 2배 크기로 setting
        new_list = [None] * new_max_size        
        self.max_size = new_max_size            # max_size 변경
        self.arrList = new_list                 # 새로운 만든 list object를 arrList attribute의 object로 설정

    def add(self,data):
        if self.pre_index >= self.max_size:
            self._increase_size()

        self.arrList[self.pre_index] = data
        self.pre_index += 1

    def insert(self,data,index):
        if self.pre_index == self.max_size:
            self._increase_size()
    
        # while version
        #tmp_index = self.pre_index
        #while tmp_index > index:
        #    self.arrList[tmp_index] = self.arrList[tmp_index-1]
        #    tmp_index -= 1
        #self.arrList[index] = data   
        #self.pre_index += 1    

        #for version
        for i in range(self.pre_index,index,-1):
            self.arrList[i] = self.arrList[i-1]
        self.arrList[index] = data   
        self.pre_index += 1    

    def delete(self,index):
        if index >= self.max_size or index < 0:
            raise Exception('Invalid index')
        
        # while version
        #while index < self.pre_index:
        #    self.arrList[index] = self.arrList[index+1] 
        #    index += 1
        #self.pre_index -= 1

        # for version
        for i in range(index, self.pre_index):
            self.arrList[i] = self.arrList[i+1]
        self.pre_index -= 1
```

```python
Arr_list = ArrayList()

Arr_list.add(1)
Arr_list.add(2)
Arr_list.add(3)
Arr_list.add(4)

print(Arr_list.arrList)
print(Arr_list.pre_index)

Arr_list.insert(5,2)
print(Arr_list.arrList)
print(Arr_list.pre_index)

Arr_list.delete(2)
print(Arr_list.arrList)
print(Arr_list.pre_index)
```
