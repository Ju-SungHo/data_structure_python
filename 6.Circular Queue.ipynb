## Circular Queue
Array list 형태로만 구현할 것 이다

Array list 형태 Linear queue의 문제점을 보완한 Queue


```python
# implement by self made method
class Circular_Queue(object):
    def __init__(self,max=11):
        self.Queue = [None] * max
        self.head=0
        self.tail=0
        self.max_size = max
        self.size=0
    
    def enqueue(self,data):
        if self.IsFull():
            return None
        self.Queue[self.head] = data
        self.head = (self.head+1)%self.max_size
        self.size += 1
    
    def dequeue(self):
        if self.IsEmpty():
            return None
        val = self.Queue[self.tail]
        self.tail = (self.tail+1)%self.max_size
        self.size -= 1
        return val

    def top(self):
        if self.IsEmpty():
            return None
        return self.Queue[self.tail]

    def sizeOf(self):
        return self.size

    def IsEmpty(self):
        if self.head == self.tail:
            return True
        return False
    
    def IsFull(self):
        if (self.head+1)% self.max_size == self.tail:
            return True
        return False
  
```

```python
# implement by list's method
class Circular_Queue(object):
    def __init__(self,max=11):
        self.Queue = []

    def __len__(self):
        return len(self.Queue)      
    
    def enqueue(self,data):
        self.Queue.append(data)
    
    def dequeue(self):
        if self.IsEmpty():
            return None
        return self.Queue.pop(0)

    def top(self):
        return self.Queue[0]

    def sizeOf(self):
        return len(self.Queue)

    def IsEmpty(self):
        return len(self.Queue) == 0        
```

```python
test = Circular_Queue()
for i in range(10):
    test.enqueue(i**2)

if test.enqueue(10**2) is None:
    print("Queue is Full") 

print(test.sizeOf())
print(test.top())

for i in range(10):
    print(test.dequeue(), end=' ')
print()

if test.dequeue() is None:
    print("Queue is Empty") 

print("====================================================")

for i in range(10):
    test.enqueue(i**3)

print(test.sizeOf())
print(test.top())

for i in range(10):
    print(test.dequeue(), end=' ')
```
