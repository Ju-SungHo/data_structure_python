```python
class Min_Heap(object):
    def __init__(self,val=[]):
        self.heap = val

    # input list에 대하여 heap 형태로 만든다.
    def make_heap(self):
        n = len(self.heap)
        for k in range(n-1 , -1, -1):
            self._heapify_down(k,n)    

    def __len__(self):
        return len(self.heap)

    def insert(self, val):
        self.heap.append(val)
        key = len(self.heap) - 1
        self._heapify_up(key)

    def delete(self):
        if len(self.heap) == 0: return None

        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self._heapify_down(0,len(self.heap))

    # Insert에서 사용
    def _heapify_up(self,k):
        parent = (k-1)//2

        if k > 0 and self.heap[k] < self.heap[parent]:
            self.swap(parent, k)
            self._heapify_up(parent)

    # make_heap과 delete에서 사용
    def _heapify_down(self,k, n):
        left = (k*2)+1
        right = (k*2)+2
        min_idx = k

        if left < n and self.heap[left] < self.heap[min_idx]:
            min_idx = left
        if right < n and self.heap[right] < self.heap[min_idx]:
            min_idx = right

        # node 값이 변경됨으로 sub-tree에 대해 다시 정렬
        if min_idx != k:
            self.swap(min_idx, k)
            self._heapify_down(min_idx,n)   
    
    def find_min(self):
        return self.heap[0]

    def swap(self, idx1, idx2):
        self.heap[idx1], self.heap[idx2] = self.heap[idx2], self.heap[idx1]
    
    def heap_sort(self,n):
        if n <= 0: return None

        self.swap(0,n-1)
        self._heapify_down(0,n-1)
        self.heap_sort(n-1)

    def clear(self):
        self.heap = []
    
```
