```python
class node(object):
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next_node = next_node

class  Linked_list(object):
    def __init__(self, head=None):
        self.head = head                    # Linked-list pointer 
        self.tail = None
        self.size = 0                        # Linked-list size

    def print(self):
        curr = self.head
        while curr != None:
            print(f"{curr.data}", end=' ')
            curr = curr.next_node
        print()

    # index 번째 list의 data return
    def get(self, index):
        curr = self.head
        i = 0
        while i < index:
            curr = curr.next_node
            i += 1
        return curr.data

    # List에 data의 존재 유무/index return
    def IsContains(self,data):  
        curr = self.head
        index = 0
        while curr != None:
            if curr.data != data:
                curr=curr.next_node
                index += 1
            else:
                return True, index
        return False, None

    # list의 size return
    def sizeOf(self):
        return self.size

    def append(self,data,mode='front'):     # mode = front, back
        if mode == 'front':
            new_node = node(data, self.head)    # new_node.data = data, new_node.next_node = self.head
            
            if self.head == None:
                self.tail = new_node

            self.head = new_node
            
        elif mode == 'back':   
            new_node = node(data)
            # 첫 node일 경우 tail과 head 모두 None이여서 tail.next_node는 비어있는 공간이기 때문에, 이 경우를 고려한 if 문 추가
            if self.tail == None:
                self.head = new_node
            else:
                self.tail.next_node = new_node
            
            self.tail = new_node    # 현재 tail의 상태와 상관없이 마지막엔 tail이 new_node를 가리키도록 설정해야 한다.
            
        else:
            print("this mode is not exist")
            return False
            
        self.size += 1

    def insert(self,data,index):
        prev = self.head
        i = 0
        while i < index-1:
            prev = prev.next_node
            i += 1

        new_node = node(data, prev.next_node)
        prev.next_node = new_node
        self.size += 1

    # 해당 data의 node delete
    def delete(self, data):
        curr = self.head
        prev = None
        while curr != None:
            if curr.data != data:
                prev = curr
                curr = curr.next_node
            else:
                if prev == None:
                    self.head = curr.next_node
                else:
                    prev.next_node = curr.next_node    
                self.size -= 1
                return
        return False

    def clear(self):
        self.size = 0
        self.head = None
        self.tail = None
```

```python
print("Test Code")
test = Linked_list()
for i in range(10):
    test.append(i**2,'back')

test.print()
print("Linked-list's size is {}".format(test.sizeOf()))
print("index 7's value = {}".format(test.get(7)))

res, idx = test.IsContains(49)
print(f"49 is contained in Linked-list? result = {res}, index = {idx}")

print("================================================================================")

test.insert(100, 4)

test.print()
print("Linked-list's size is {}".format(test.sizeOf()))
print("index 7's value = {}".format(test.get(7)))

res, idx = test.IsContains(49)
print(f"49 is contained in Linked-list? result = {res}, index = {idx}")

print("================================================================================")

test.delete(100)
test.print()
print("Linked-list's size is {}".format(test.sizeOf()))
print("index 7's value = {}".format(test.get(7)))

res, idx = test.IsContains(500)
print(f"49 is contained in Linked-list? result = {res}, index = {idx}")

print("================================================================================")
test.clear()

for i in range(11):
    test.append(i**3,'back')

test.print()
print("Linked-list's size is {}".format(test.sizeOf()))
print("index 7's value = {}".format(test.get(7)))

res, idx = test.IsContains(49)
print(f"49 is contained in Linked-list? result = {res}, index = {idx}")
```
